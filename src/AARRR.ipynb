> RFM 분석을 통해 고객의 거래 데이터 기반으로
> 
> 
> “누가 최근에, 얼마나 자주, 얼마나 많이” 활동했는지를 탐색했습니다.
> 
> 이제 이를 기반으로 **AARRR 퍼널 분석(획득–활성–유지–수익–추천)** 으로 넘어가,
> 
> 고객이 어떤 경로를 통해 유입되고, 얼마나 지속적으로 참여하며,
> 
> 최종적으로 어떤 가치를 만들어내는지 단계별로 살펴보겠습니다.
>

> 다음 단계에서는 사용자 흐름 전반을 AARRR 모델에 따라 분석합니다.
> 
> 
> 가입(`signup_step`), 첫 거래(`isActivation`), 재방문 빈도, 결제 금액, 추천(`isReferral`) 등의 변수를 활용해
> 
> 각 단계별 전환율과 유지율을 파악하고,
> 
> RFM에서 확인된 **핵심 활동 고객군이 어떤 경로로 수익을 창출하는지** 연결해 해석할 예정입니다.
>
